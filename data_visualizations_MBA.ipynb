# Countplot 

```python
import matplotlib.pyplot as plt  # Importing matplotlib for plotting
import seaborn as sns  # Importing seaborn for data visualization

plt.figure(figsize=(6, 6))  # Set the size of the plot to 6x6 inches
sns.countplot(data=df, x='gender')  # Create a count plot for the 'gender' column in the dataset
plt.title('Gender Distribution')  # Add a title to the plot
plt.xlabel('Gender')  # Set the label for the x-axis
plt.ylabel('Count')  # Set the label for the y-axis
plt.show()  # Display the plot
```

```python
plt.figure(figsize=(8, 6))  # Set the size of the plot to 8x6 inches
sns.countplot(data=df, x='international')  # Create a count plot for the 'international' column in the dataset
plt.title('International vs. Domestic Applicants')  # Add a title to the plot
plt.xlabel('International')  # Set the label for the x-axis
plt.ylabel('Count')  # Set the label for the y-axis
plt.show()  # Display the plot

```

# Histogram Plot

```python
plt.figure(figsize=(8, 6))  # Set the size of the plot to 8x6 inches
sns.countplot(data=df, x='race')  # Create a count plot for the 'race' column in the dataset
plt.title('Racial Distribution')  # Add a title to the plot
plt.xlabel('Race')  # Set the label for the x-axis
plt.ylabel('Count')  # Set the label for the y-axis
plt.show()  # Display the plot

```

```python
plt.figure(figsize=(12, 8))  # Set the size of the plot to 12x8 inches
sns.countplot(data=df, x='major')  # Create a count plot for the 'major' column in the dataset
plt.title('Distribution of Undergraduate Majors')  # Add a title to the plot
plt.xlabel('Major')  # Set the label for the x-axis
plt.ylabel('Count')  # Set the label for the y-axis
plt.show()  # Display the plot

```

```python
plt.figure(figsize=(8, 6))  # Set the size of the plot to 8x6 inches
sns.histplot(data=df, x='gpa', bins=20, kde=True)  # Create a histogram for 'gpa' with 20 bins and a KDE curve
plt.title('GPA Distribution')  # Add a title to the plot
plt.xlabel('GPA')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis
plt.show()  # Display the plot
```

# Plotting the distribution of GMAT scores with a KDE curve


```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='gmat',bins=10,kde=True)
plt.title('GMAT Score Distribution')
plt.xlabel('GMAT Score')
plt.ylabel('Frequency')
plt.show()
```

# Plotting the distribution of work experience with a KDE curve


```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='work_exp',bins=10,kde=True)
plt.title('Work Experience Distribution')
plt.xlabel('Years of Work Experience')
plt.ylabel('Frequency')
plt.show()
```

# Plotting the count of occurrences for each work industry


```python
plt.figure(figsize=(12, 12))
sns.countplot(data=df, x='work_industry')
plt.title('Distribution of Work Industry')
plt.ylabel('Count')
plt.xlabel('Industry')
plt.show()
```

# Plotting the count of admission outcomes


```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='admission')
plt.title('Admission Outcomes')
plt.xlabel('Admission Status')
plt.ylabel('Count')
plt.show()
```

# Heatmap for Correlation

```python
plt.figure(figsize=(10, 8))
corr = df[['gpa', 'gmat', 'work_exp']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

# Line Graph

```python
# Line Graph 
plt.figure(figsize=(10, 12))
plt.plot(df['application_id'], df['gpa'], marker='o')
plt.title('GPA over Application IDs')
plt.xlabel('Application ID')
plt.ylabel('GPA')
plt.show()
```

# Scatter Plot

```python
# Scatter Plot 
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='gmat', y='gpa',hue='admission')
plt.title('GMAT Scores vs. GPA')
plt.xlabel('GMAT Score')
plt.ylabel('GPA')
plt.show()
```

# Pair Plot

```python

sns.pairplot(df[['gpa', 'gmat', 'work_exp', 'admission']], hue='admission')
plt.suptitle('Pair Plot of GPA, GMAT, Work Experience, and Admission', y=1.02)
plt.show()
```

# Box Plot

```python
fig = plt.figure(figsize =(10, 7))
sns.boxplot(df[['gpa','work_exp','gmat']])
plt.title('Box plot')
plt.show()
```

# Pie Chart

```python
gender_counts = df['gender'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Gender Distribution')
plt.axis('equal')  
plt.show()
```

```python
international_counts = df['international'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(international_counts, labels=['Domestic', 'International'], autopct='%1.1f%%')
plt.title('International vs Domestic Applicants')
plt.axis('equal')
plt.show()
```
